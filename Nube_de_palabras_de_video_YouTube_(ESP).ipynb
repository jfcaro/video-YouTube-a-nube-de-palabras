```python
# Cargar liberías

from youtube_transcript_api import YouTubeTranscriptApi
from wordcloud import WordCloud
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words_sp = set(stopwords.words('spanish'))
import matplotlib.pyplot as plt
```

```python
# directorio para guardar nube de palabas
path_Wordcloud ="./"

# otras palabras que queremos eliminar de la nube
otras = ["digamos", "puede", "hacer","vamos","cosa","entonces","creo","totalmente", "pues", "bueno","si","escena","toda", "aquí", "ahora", "vale","podemos","ver","solamente","fijaos","cosas", "luego", "música", "haciendo","verdad", "hace"]

# Actualizamos la lista de STOPWORDS con las que proporciona nltk+otras
STOPWORDS = list(stop_words_sp)+ otras
```

```python
# Función que devuelve la transcripción de un video Youtube

def get_transcript (link):
  id = link.split("https://www.youtube.com/watch?v=")[1]
  transcripcion = YouTubeTranscriptApi.get_transcript(id, languages=['es'])
  transcript = ""
  for i in range(0,len(transcripcion)):
    transcript +=str(transcripcion[i]['text'])+" "
  return transcript  

# Función que genera la nube de palabras de la trasncripción 

def gen_wordcloud(transcription,index):
  wordcloud = WordCloud(stopwords=STOPWORDS, background_color='white',width=2400,height=2000).generate(transcription)
  plt.figure(figsize=(12,15))
  plt.imshow(wordcloud)
  plt.axis('off')
  plt.savefig(path_Wordcloud +str(index)+".png")
  plt.show()
  

# Función que muestra la nuve de palabras

def show_transcript (transcripcion):
  palabras = transcripcion.split()
  renglon =""
  for i in range(0,len(palabras)):
    renglon = renglon + palabras[i]+" "
    if i % 40 ==0:
      print (renglon)
      renglon =""
```

```python
video = input("introduce url del tipo https://www.youtube.com/watch?v=XXXXXXX ")
```

```python
print ("************  MOSTRANDO LA TRANSCRIPCIÓN  ************")
print ("video: ", video )
transcripcion = get_transcript(video)
print ("\n\n")
show_transcript(transcripcion)
```

```python
print ("************  GENERANDO NUBE DE PALABRAS  ************")
print ("video: ", video )
print ("stop_words: ",stop_words_sp)
print ("palabras añadidas a stop_words:", otras)
print ("\n")
gen_wordcloud(transcripcion,"img")
print ("\n\n fichero img.png con nube de palabras creado en: ",path_Wordcloud )
```
